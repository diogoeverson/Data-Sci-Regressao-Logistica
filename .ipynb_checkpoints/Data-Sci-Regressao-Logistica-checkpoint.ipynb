```python
#Regressão logística é um modelo para classificação apesar de conter em seu nome regressão
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib as mpl
from sklearn.pipeline import make_pipeline
from sklearn.preprocessing import StandardScaler
from sklearn.linear_model import LogisticRegression
from sklearn.model_selection import train_test_split
```

```python
df = pd.read_csv('automobile_data.csv')
```

```python
df.head(50)
```

```python
#variável target
df['aspiration'].unique()
```

```python
df.columns
```

```python
#movendo a variável target para a extrema direita do dataset
column_names = ['symboling', 'normalized-losses', 'make', 'fuel-type',
       'num-of-doors', 'body-style', 'drive-wheels', 'engine-location',
       'wheel-base', 'length', 'width', 'height', 'curb-weight', 'engine-type',
       'num-of-cylinders', 'engine-size', 'fuel-system', 'bore', 'stroke',
       'compression-ratio', 'horsepower', 'peak-rpm', 'city-mpg',
       'highway-mpg', 'price', 'aspiration']

df = df.reindex(columns=column_names)
```

```python
df.columns
```

```python
#separando variáveis preditoras da target
X = df.iloc[:,:-1]
y = df['aspiration'].values
```

```python
#loop que percorre lista de colunas categóricas que terão valores por categoria convertidas em número 
X = X.astype(str)
novo_valor_numerico = 0
i = 0
for x in X.columns:
    for z in X[x].drop_duplicates():        
        X[x].replace({z: str(i)}, inplace=True)
        i+=1
    i = 0
X = X.astype(int)
```

```python
X.head()
```

```python
y
```

```python
# Divisão em dados de treino e de teste
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
# Cria o modelo
modelo = LogisticRegression()
```

```python
modelo.fit(X_train, y_train)
```

```python
predictions = modelo.predict(X_test)
```

```python
modelo.score(X_test, y_test)
```

```python
#Precisão, acurácia e f1-score próximos de 90%
from sklearn.metrics import classification_report
print(classification_report(y_test,predictions))
```
